## Introduction  
This project analyzes the Amazon Electronics dataset containing product ratings, brands, categories, and user data.  
The goal is to understand product popularity, user activity, and rating trends.


```python
from google.colab import files
import pandas as pd
import io, zipfile
import seaborn as sns
import matplotlib.pyplot as plt

# Step A: Upload the dataset
uploaded = files.upload()

# Step B: Handle ZIP or CSV
filename = list(uploaded.keys())[0]

if filename.endswith(".zip"):
    with zipfile.ZipFile(io.BytesIO(uploaded[filename]), 'r') as zip_ref:
        zip_ref.extractall()
        csv_file = zip_ref.namelist()[0]
        df = pd.read_csv(csv_file)
else:
    df = pd.read_csv(io.BytesIO(uploaded[filename]))

# Step C: Preview the dataset
df.head()

```

```python
print("Before cleaning:", df.shape)
print("\nMissing values:\n", df.isnull().sum())

# Drop nulls and duplicates
df = df.dropna()
df = df.drop_duplicates()

# Standardize text columns (correct column names)
df['brand'] = df['brand'].str.strip().str.lower()
df['category'] = df['category'].str.strip().str.lower()

print("\nAfter cleaning:", df.shape)
df.head()
```

```python
sns.countplot(x='rating', data=df)
plt.title("Distribution of Ratings")
plt.show()
```

"most ratings are between 4 and 5 stars, showing general customer satisfaction."

```python
df['brand'].value_counts().head(10).plot(kind='bar', figsize=(10,5))
plt.title("Top 10 Brands by Product Count")
plt.show()
```

"logitech,bose and sony are the top brands in the dataset."

```python
df['category'].value_counts().head(10).plot(kind='barh', figsize=(10,5))
plt.title("Top 10 Categories by Count")
plt.show()
```

"most products belong to headphones,computers and accessories

```python
df.groupby('year')['rating'].mean().plot(marker='o')
plt.title("Average Rating by Year")
plt.show()
```

Insight from the graph:

From around 2000 to 2006, the average rating fluctuated (went up and down).

After that, from 2006 onwards, the trend mostly increases, meaning customers gave better average ratings over time.

Theres a dip towards the very end, suggesting that the average rating slightly decreased in recent years

## Conclusion  

- Most ratings are positive (45 stars).  
- logitech,bose and sony dominate the brand list.  
- headphones,computers and accessories are the top categories.  
- We can see that the average rating fluctuates year by year.

Overall, there is a slight upward trend in ratings, meaning that over time, customers are giving better ratings to electronics products.

However, there is a drop in average rating in the final year, which could suggest a decrease in product satisfaction or an increase in negative reviews.
